* Lets try another approach and use LLM model as OCR, we will compare the results and see.



```python
import google.generativeai as genai
import os
```

```python
key=os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=key)
```

```python
model=genai.GenerativeModel("gemini-2.5-flash-lite")
```

```python
prompt='''You act as an OCR application on the given image and extract the text from it.
Give only the text as output and do not give any explanation or description.
'''
```

```python
# lets load and process the image.
img=cv2.imread('Capture1.png')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 
image_grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,image_bw=cv2.threshold(image_grey,90,255,cv2.THRESH_BINARY) # To convert to black and white
```

```python
# THe image that CV2 reads is in numpy array format, we need to convert it to image object.
from PIL import Image
final_image=Image.fromarray(image_bw)
```

```python
final_image
```

```python
response=model.generate_content([prompt,final_image])
output_text=response.text
print(output_text)
```

Note:

OCR based on LLM model seems to be a better option to extract text even from handwritten notes. It may not be able to extract 100% but still giving better output than tesseract
